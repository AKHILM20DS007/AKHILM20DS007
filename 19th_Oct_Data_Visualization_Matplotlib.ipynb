```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/employees.csv")
data.head()
```

```python
### Line Graph
salary = data.iloc[0:50]['Salary']
index = data.index[0:50]
plt.plot(index,salary,color="red",lw=3,linestyle="dotted")  ## solid,dashed,dotted
plt.show()
```

```python
### Scatter Plot Graph
salary = data.iloc[0:50]['Salary']
index = data.index[0:50]
### *,o,d,v,<,>,^,+
plt.scatter(index,salary,color="red",s=150,marker="*")
plt.show()
```

```python
data['First Name'] = data['First Name'].astype("str")
```

```python
### Bar Plot Graph
salary = data.iloc[0:30]['Salary']
name = data.iloc[0:30]['First Name']
plt.xticks(rotation=90)
plt.bar(name,salary,color="orange")
plt.show()
```

```python
### Horizontal Bar Plot Graph
salary = data.iloc[0:30]['Salary']
name = data.iloc[0:30]['First Name']
plt.barh(name,salary,color="orange")
plt.show()
```

```python
teams = pd.DataFrame(data['Team'].value_counts())
teams
```

```python
teams = teams.reset_index()
```

```python
teams.columns = ["Team","Count"]
```

```python
plt.figure(figsize=(6,6))
a = (0,0,0.3,0,0,0,0,0,0,0.5)
plt.pie(x=teams['Count'],labels=teams['Team'],autopct="%0.2f%%",explode=a)
plt.show()
```

```python
### Bar Plot Graph
salary = data.iloc[0:30]['Salary']
name = data.iloc[0:30]['First Name']
plt.figure(figsize=(8,5))
plt.title("Salary Graph",color="blue",fontsize=15)
plt.xlabel("Employee Name",color="purple",fontsize=14)
plt.ylabel("Salary",color="purple",fontsize=14)
plt.xticks(rotation=90)
plt.bar(name,salary,color="orange",label="Salary")
plt.legend()
plt.savefig("salarygraph.png")
plt.show()
```
