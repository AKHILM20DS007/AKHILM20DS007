```python
import cv2
import matplotlib.pyplot as plt
```

### Face Detection

```python
img = cv2.imread('face.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier('face.xml')
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
plt.figure(figsize=(6,10))
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```

### Face and Eyes Detection

```python
import cv2
cam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier('face.xml')
eye_cascade = cv2.CascadeClassifier('eye.xml')
while True:
    status,frames = cam.read()
    faces = face_cascade.detectMultiScale(frames,1.12)
    for x,y,w,h in faces:
        cv2.rectangle(frames,(x,y),(x+w,y+h),(255,0,0),3)
        single_face = frames[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(single_face,1.22)
        for ex,ey,ew,eh in eyes:
            cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(0,0,255),3)
    cv2.imshow("My Webcam",frames)
    if cv2.waitKey(1) == 13:   ## ASCII Code for enter 
        break
cam.release()
cv2.destroyAllWindows()
```

### Object Detection by Color (Blue)

```python
import cv2
import numpy as np
cam = cv2.VideoCapture(0)
while True:
    status,frames = cam.read()
    hsv = cv2.cvtColor(frames,cv2.COLOR_BGR2HSV)
    low = np.array([[90,80,10]])
    high = np.array([[120,255,255]])
    mask_image = cv2.inRange(hsv,low,high)
    final = cv2.bitwise_and(frames,frames,mask=mask_image)
    cv2.imshow("Final Image",final)
    cv2.imshow("Mask Image",mask_image)
    cv2.imshow("HSV",hsv)
    cv2.imshow("My Webcam",frames)
    if cv2.waitKey(1) == 13:   ## ASCII Code for enter 
        break
cam.release()
cv2.destroyAllWindows()
```

#### Blue Screen Application

```python
import cv2
import numpy as np
cam = cv2.VideoCapture(0)
while True:
    status,frames = cam.read()
    back_image = cv2.imread('flower.jpg')
    frames = cv2.resize(frames,(640,480))
    back_image = cv2.resize(back_image,(640,480))
    hsv = cv2.cvtColor(frames,cv2.COLOR_BGR2HSV)
    low = np.array([[90,80,10]])
    high = np.array([[120,255,255]])
    mask_image = cv2.inRange(hsv,low,high)
    final = cv2.bitwise_and(frames,frames,mask=mask_image)
    result = frames - final
    result = np.where(result==0,back_image,result)
    cv2.imshow("Result",result)
    cv2.imshow("Final Image",final)
    cv2.imshow("Mask Image",mask_image)
    cv2.imshow("HSV",hsv)
    cv2.imshow("My Webcam",frames)
    if cv2.waitKey(1) == 13:   ## ASCII Code for enter 
        break
cam.release()
cv2.destroyAllWindows()
```
